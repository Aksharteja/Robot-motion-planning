## 1. Bug Algorithms (Bug2)
This task demonstrates the Bug2 algorithm, where the robot moves towards the goal but navigates around obstacles when encountered.

```python
import matplotlib.pyplot as plt
import numpy as np

madhya = np.array([8, 4])
vyasa = 3
aarambham = np.array([0, 0])
lakshyam = np.array([12, 10])
dari = [aarambham.copy()]

padugupudu = 0.5
sahishnuta = 0.3
gatimettu = 1000

rekha_disha = (lakshyam - aarambham) / np.linalg.norm(lakshyam - aarambham)
prasthutam = aarambham.copy()
sparsha = None
vichheda = None
stithi = 'lakshyam_ku_veladam'

def rekha_pai_unda(iti):
    prarambha_diff = iti - aarambham
    lambam = np.dot(prarambha_diff, rekha_disha)
    bindu = aarambham + lambam * rekha_disha
    dhooram = np.linalg.norm(iti - bindu)
    return dhooram < sahishnuta

def chetta_lakshyam(iti):
    return np.linalg.norm(iti - lakshyam) < padugupudu

def lakshyam_dhooram(iti):
    return np.linalg.norm(iti - lakshyam)

def pracheeksha_sutram(iti, kona_padam=10):
    veekshanam = iti - madhya
    prachina_kona = np.arctan2(veekshanam[1], veekshanam[0])
    kotta_kona = prachina_kona + np.deg2rad(kona_padam)
    kotta_iti = madhya + vyasa * np.array([np.cos(kotta_kona), np.sin(kotta_kona)])
    return kotta_iti

for _ in range(gatimettu):
    if chetta_lakshyam(prasthutam):
        break

    d_veekshanam = lakshyam - prasthutam
    veekshanam_dhooram = np.linalg.norm(d_veekshanam)
    disha = d_veekshanam / veekshanam_dhooram

    if stithi == 'lakshyam_ku_veladam':
        kotta_iti = prasthutam + padugupudu * disha
        pramadam_dhooram = np.linalg.norm(kotta_iti - madhya)

        if pramadam_dhooram <= vyasa + sahishnuta:
            sparsha = prasthutam.copy()
            stithi = 'pareetham_chuttu_veladam'
            sneham = prasthutam.copy()
            thakkuva_dhooram = lakshyam_dhooram(prasthutam)
        else:
            prasthutam = kotta_iti
            dari.append(prasthutam.copy())

    elif stithi == 'pareetham_chuttu_veladam':
        kotta_iti = pracheeksha_sutram(prasthutam)
        prasthutam = kotta_iti
        dari.append(prasthutam.copy())

        ippudu_dhooram = lakshyam_dhooram(prasthutam)
        if ippudu_dhooram < thakkuva_dhooram:
            thakkuva_dhooram = ippudu_dhooram
            sneham = prasthutam.copy()

        if rekha_pai_unda(prasthutam) and lakshyam_dhooram(prasthutam) < lakshyam_dhooram(sparsha):
            stithi = 'lakshyam_ku_veladam'

dari = np.array(dari)
fig, aksharam = plt.subplots()
badha = plt.Circle(madhya, vyasa, color='gray', alpha=0.5)
aksharam.add_patch(badha)
plt.plot(dari[:, 0], dari[:, 1], marker='o', label='Dari')
plt.plot([aarambham[0]], [aarambham[1]], 'go', label='Start')
plt.plot([lakshyam[0]], [lakshyam[1]], 'ro', label='Goal')
plt.plot([aarambham[0], lakshyam[0]], [aarambham[1], lakshyam[1]], 'k--', label='M-line')
plt.title('Bug2 Algorithm')
plt.axis('equal')
plt.grid(True)
plt.legend()
plt.show()
```
