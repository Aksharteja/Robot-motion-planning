# 2. Configuration Space Exploration




This practical explores configuration space by inflating obstacles based on robot size.





```python
import matplotlib.pyplot as plt

bandhakalu = [((2, 2), (3, 4)), ((5, 5), (7, 7))]
yantra_radius = 0.5

def vistharimpabadhakalu(praantaralu, re):
    vistharita = []
    for (x1, y1), (x2, y2) in praantaralu:
        vistharita.append(((x1 - re, y1 - re), (x2 + re, y2 + re)))
    return vistharita

chitra, aksharam = plt.subplots()
for (x1, y1), (x2, y2) in bandhakalu:
    aksharam.add_patch(plt.Rectangle((x1, y1), x2 - x1, y2 - y1, color='gray'))

vistharita_badhakalu = vistharimpabadhakalu(bandhakalu, yantra_radius)
for (x1, y1), (x2, y2) in vistharita_badhakalu:
    aksharam.add_patch(plt.Rectangle((x1, y1), x2 - x1, y2 - y1, color='red', alpha=0.3))

plt.title("Parinama Sthalam (Configuration Space)")
plt.grid(True)
plt.axis('equal')
plt.show()
```
